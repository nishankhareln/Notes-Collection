| **Feature**         | **White Noise**                              | **Random Walk**                                                               |
| ------------------- | -------------------------------------------- | ----------------------------------------------------------------------------- |
| **Stationarity**    | Stationary (constant mean, variance, ACF)    | Non-stationary (mean and variance change over time)                           |
| **Autocorrelation** | No autocorrelation at any lag (except lag 0) | Strong autocorrelation, especially at shorter lags                            |
| **Predictability**  | Unpredictable (best forecast is its mean)    | Short-term predictable (best forecast is last value), long-term unpredictable |
| **Dependence**      | Independent observations (memoryless)        | Each value depends on the previous value (has memory)                         |
| **Origin**          | The "shock" or "innovation" term             | The cumulative sum of white noise                                             |
